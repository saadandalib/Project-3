```python
import pandas as pd
```

```python
file = 'yelp_business.csv'
```

```python
yelp_df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
yelp_df.head()
```

```python
yelp_df = yelp_df[["name","address","city","state", "postal_code", "latitude","longitude", "stars", "review_count", "categories"]]
```

```python
yelp_df.head()
```

```python
yelp_df.count()
```

```python
cleaned_yelp_df = yelp_df.dropna(how="any")
```

```python
cleaned_yelp_df.count()
```

```python
cleaned_yelp_df.to_csv(r'yelp_business.csv',index=False)
```

```python
cleaned_yelp_df.head()
```

```python
is_az = cleaned_yelp_df['state']=='AZ'
az_yelp_df = cleaned_yelp_df[is_az]
```

```python
az_yelp_df.head()
```

```python
az_yelp_df.to_csv(r'az_yelp_business.csv',index=False)
```
